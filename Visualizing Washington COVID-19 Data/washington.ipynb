## Step 1: Import and inspect

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('datasets/washington-covid.csv')
df.head()
```

## Step 2: Cleaning up a bit
It does seem that we could take advantage of the date column as our x-axis.

But first, let us convert its type to `datetime` and then sort it in an ascending order.

```python
df.date = pd.to_datetime(df.date, format='%Y%m%d')
df = df.sort_values('date', ignore_index=True)
df.head()
```

You may now ~~kiss the bride~~ assign the x-axis value.

```python
x = df.date
```

As our y-axis, we will be using the `deathIncrease` column for now.

```python
y = df.deathIncrease.cumsum()
```

## Step 3: "Let there be visualization!"

```python
plt.style.use('seaborn')
plt.plot(x, y)

plt.title('Number of Death Increase in WDC')
plt.xlabel('Date')
plt.ylabel('No. Deaths')
plt.xticks(rotation=45)
```

## Step 4: Sanity check!
The shape of our plot shooould be the same if we use the `death` column as our y-axis instead.

```python
y = df.death

plt.plot(x, y)

plt.title('Number of Deaths in WDC')
plt.xlabel('Date')
plt.ylabel('No. Deaths')
plt.xticks(rotation=45)
```

It appears that all is in order!
